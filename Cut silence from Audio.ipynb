```python
from pyAudioAnalysis import audioBasicIO as aIO
```

```python
from pyAudioAnalysis import audioSegmentation as aS
```

```python
import subprocess
```

```python
[Fs, x] = aIO.readAudioFile("./wav/teste3.wav")
```

```python
segments = aS.silenceRemoval(x, Fs, 0.010, 0.010, smoothWindow = 0.5, Weight = 0.3, plot = True)
```

```python
for seg in segments:
    print "Noise = ", seg
```

```python
startx, endx = segments[1]
```

```python
commandox = "play ./wav/teste3.wav"
```

```python
subprocess.call(commandox, shell=True)
```
